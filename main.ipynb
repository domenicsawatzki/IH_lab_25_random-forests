```python
categorical_df = pd.read_csv('files_for_lab/categorical.csv')
numerical_df = pd.read_csv('files_for_lab/numerical.csv')
target_df = pd.read_csv('files_for_lab/target.csv')
```

```python

X = pd.concat([categorical_df, numerical_df], axis = 1)
y = target_df['TARGET_B']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1337)
```

```python
X_train_numerical = X_train.select_dtypes(np.number)
X_test_numerical = X_test.select_dtypes(np.number)

X_train_categorical = X_train.select_dtypes(object)
X_test_categorical = X_test.select_dtypes(object)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler().fit(X_train_numerical)

X_train_numerical_scaled = scaler.transform(X_train_numerical)
X_train_numerical_scaled_df = pd.DataFrame(X_train_numerical_scaled, columns=scaler.get_feature_names_out(X_train_numerical.columns))

X_test_numerical_scaled = scaler.transform(X_test_numerical)
X_test_numerical_scaled_df = pd.DataFrame(X_test_numerical_scaled, columns=scaler.get_feature_names_out(X_test_numerical.columns))
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first')
encoder.fit(X_train_categorical)

X_train_categorical_encoded = encoder.transform(X_train_categorical)
X_train_categorical_encoded_df = pd.DataFrame(X_train_categorical_encoded.toarray(), columns=encoder.get_feature_names_out(X_train_categorical.columns))

X_test_categorical_encoded = encoder.transform(X_test_categorical)
X_test_categorical_encoded_df = pd.DataFrame(X_test_categorical_encoded.toarray(), columns=encoder.get_feature_names_out(X_test_categorical.columns))
```

```python
X_train_transformed = pd.concat([X_train_categorical_encoded_df, X_train_numerical_scaled_df], axis = 1)
display(X_train_transformed)

X_test_transformed = pd.concat([X_test_categorical_encoded_df, X_test_numerical_scaled_df], axis = 1)
X_test_transformed
```

```python
X_train_transformed = X_train_transformed.reset_index(drop=True)

print(X_train_transformed.isna().sum().sum())
X_test_transformed = X_test_transformed.reset_index(drop=True)
```

```python
y_train = y_train.reset_index(drop=True)
imbalance_data = pd.concat([X_train_transformed, y_train], axis = 1)

imbalance_data['TARGET_B'].value_counts()
imbalance_data.isna().sum().sum()
```

```python
category_0 = imbalance_data[imbalance_data['TARGET_B'] == 0]
category_1 = imbalance_data[imbalance_data['TARGET_B'] == 1]

print(category_0.shape)
print(category_1.shape)
```

```python
from sklearn.utils import resample
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0))

data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)

print(data_upsampled['TARGET_B'].value_counts())
display(data_upsampled)
```

```python
X_train_up = data_upsampled.drop('TARGET_B', axis = 1)
y_train_up = data_upsampled['TARGET_B']
```

```python
X_train_up
```

```python
y_train_up = y_train_up.reset_index(drop=True)
```

```python
X_train_up 
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

# Adjusted hyperparameters
RFC = RandomForestClassifier(max_depth=5, 
                             min_samples_split=20, 
                             min_samples_leaf=20, 
                             max_samples=None,  # Adjusted to None
                             random_state=42) 

from sklearn.model_selection import cross_val_score


mean_score = np.mean(cross_val_score(RFC, X_train_up, y_train_up, cv=5)) 
RFC.fit(X_train_up, y_train_up)

# Scoring
RFC_score_train = RFC.score(X_train_up, y_train_up)
RFC_score_test = RFC.score(X_test_transformed, y_test)

print (f'The cross validation score is: {mean_score.round(3)}')
print(f'Training score: {RFC_score_train.round(3)}')
print(f'Test score: {RFC_score_test.round(3)}')

# Confusion Matrix
y_pred = RFC.predict(X_test_transformed)
print(confusion_matrix(y_test, y_pred))
```

```python

from sklearn.feature_selection import RFE
from sklearn import linear_model
lm = linear_model.LinearRegression()
rfe = RFE(lm, n_features_to_select=None, verbose=False)
rfe.fit(X_train_up, y_train_up)
```

```python
df = pd.DataFrame(data = rfe.ranking_, columns=['Rank'])
df['Column_name'] = pd.DataFrame(X_train_up).columns
test = df[df['Rank']==1]
keep_columns = []
keep_columns = test['Column_name'].tolist()
```

```python
print(X_train_up.shape)
X_train_up_features_selected = X_train_up[keep_columns]
print(X_train_up_features_selected.shape)


print(X_test_transformed.shape)
X_test_transformed_features_selected = X_test_transformed[keep_columns]
print(X_test_transformed_features_selected.shape)
```

```python
RFC = RandomForestClassifier(max_depth=5, 
                             min_samples_split=20, 
                             min_samples_leaf=20, 
                             max_samples=None,  # Adjusted to None
                             random_state=42) 

from sklearn.model_selection import cross_val_score


mean_score = np.mean(cross_val_score(RFC, X_train_up_features_selected, y_train_up, cv=5)) 
RFC.fit(X_train_up_features_selected, y_train_up)

# Scoring
RFC_score_train = RFC.score(X_train_up_features_selected, y_train_up)
RFC_score_test = RFC.score(X_test_transformed_features_selected, y_test)

print (f'The cross validation score is: {mean_score.round(3)}')
print(f'Training score: {RFC_score_train.round(3)}')
print(f'Test score: {RFC_score_test.round(3)}')

# Confusion Matrix
y_pred = RFC.predict(X_test_transformed_features_selected)
print(confusion_matrix(y_test, y_pred))
```
